```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('titanic_train.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
df['Age'].shape
```

```python
si = SimpleImputer(strategy='mean')
df['Age'] = si.fit_transform(df['Age'].values.reshape(-1,1))
```

```python
le = LabelEncoder()
df['Embarked'] = le.fit_transform(df['Embarked'].astype('str'))
```

```python
df
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
X = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
classifiers_ = [
    ("AdaBoost",AdaBoostClassifier()),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Linear SVM", SVC(kernel="linear",probability=True)),
    ("Naive Bayes",GaussianNB()),
    ("Nearest Neighbors",KNeighborsClassifier(3)),
    ("Random Forest",RandomForestClassifier(n_jobs=2)),
    ]
```

```python
clf_names = []
train_scores = []
test_scores = []
for n,clf in classifiers_:
    clf_names.append(n)
    clf = Pipeline([('POLY', poly),('CLF',clf)])

    clf.fit(X_train, y_train)
    
    train_scores.append(clf.score(X_train, y_train))
 
    test_scores.append(clf.score(X_test, y_test))
    print("---")
```

```python
plt.title('Accuracy Training Score')
plt.grid()
plt.plot(train_scores,clf_names)
plt.show()

plt.title('Accuraccy Test Score')
plt.grid()
plt.plot(test_scores,clf_names)
plt.show()
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
rfcpred = rfc.predict(X_test)
rfcpred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,rfcpred)
```

```python
print(classification_report(y_test,rfcpred))
```

```python
print(confusion_matrix(y_test,rfcpred))
```
